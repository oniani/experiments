```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("ggplot")
```

```python
SIZE: int = 8
SIGNAL: np.ndarray = np.random.uniform(low=-SIZE, high=SIZE, size=SIZE)
POSITIONS: np.ndarray = np.random.choice(range(2 * SIZE), size=SIZE, replace=False)
POSITIONS
```

```python
plt.stem(POSITIONS, SIGNAL)
plt.title("Signal x[n]")
plt.show()
```

```python
plt.stem(-POSITIONS, SIGNAL)
plt.title("Signal x[-n]")
plt.show()
```

```python
SIGNAL_INVERSE = [y for x, y in sorted(zip(-POSITIONS, SIGNAL))]
plt.stem(SIGNAL_INVERSE)
plt.title("Signal x[-n] (Without Silence)")
plt.show()
```

```python
# For even signals, x[n] = x[-n] must hold
SIGNAL_EVEN: np.ndarray = (SIGNAL + SIGNAL_INVERSE) / 2
plt.stem(SIGNAL_EVEN)
plt.title("Even Signal = (x[n] + x[-n]) / 2")
plt.show()
```

```python
# For odd signals, x[n] = -x[-n] must hold
SIGNAL_ODD: np.ndarray = (SIGNAL - SIGNAL_INVERSE) / 2
plt.stem(SIGNAL_ODD)
plt.title("Odd Signal = (x[n] - x[-n]) / 2")
plt.show()
```
